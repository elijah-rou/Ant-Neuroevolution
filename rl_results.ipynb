```python
import numpy as np
with open("reinforce.npy", "rb") as f:
    results = np.load(f)
```

```python
import matplotlib.pyplot as plt
plt.plot(results)
```

```python
import pandas as pd
scores = pd.Series(results[3263:-263])
```

```python
plt.title(f"Rolling average of reward over 100 runs; gamma=0.99, lr=0.001")
plt.xlabel("Epoch")
plt.xticks(np.arange(0, 3000+100, 500))
plt.ylabel("Rolling average reward over 100 runs")
plt.plot(scores.rolling(100).mean())
plt.savefig(f"rl_discreteAnt.png")
plt.show()
```
